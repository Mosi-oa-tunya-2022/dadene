```python
# pip install pandas_datareader
import os
import requests
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
import datetime as dt
```

The datetime() is straight forward. dt.datetime() takes date in this format dt.datetime(2018,1,1)
 and the end date can be defined as follows if you need current date. end =dt.datetime.now()

```python
start = dt.datetime(2018,1,1)
end =dt.datetime.now()
```

```python
bitcoin = web.DataReader('BTC-USD', 'yahoo', start, end)
bitcoin
```
